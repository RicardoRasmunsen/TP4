```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

ganancias = pd.read_csv('company_sales_data.csv')
ganancias
```

```python
ganancias_por_mes = ganancias[['month_number',  'total_profit']]
ganancias_por_mes
```

```python
dias_de_mes = ganancias_por_mes.month_number
total_ganancias = ganancias_por_mes.total_profit
plt.pie(total_ganancias, labels=dias_de_mes, autopct='%0.1f %%')
plt.show
```

```python
venta_productos = ganancias[['month_number', 'facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo','moisturizer']]
venta_productos=venta_productos.set_index('month_number')
venta_productos
```

```python
cambio_valores_facewash = venta_productos.facewash[0:12] = [2100, 3500, 150, 280, 920, 4560, 780, 6789, 7905, 1230, 2230, 4538]
venta_productos
```

```python
# Creamos la figura.
plt.figure(figsize=(12, 8))

# Itera a travs de las columnas y traza cada columna en el mismo grfico
for column in venta_productos.columns:
    if column != 'month_number': # Este if es para que no tome como valores para el grfico,
                                 #la columna de meses que se utilizan como indices.
      plt.plot(venta_productos.index, venta_productos[column], label=column)

# Agrega etiquetas y leyenda
plt.xlabel('mes')
plt.ylabel('cantidad')
plt.grid(True) # Muestra la grilla.
plt.legend()

# Muestra el grfico
plt.show()
```

```python

import matplotlib.pyplot as plt

# Crea una figura y un conjunto de subtramas (subplots) con una fila por elemento
fig, axs = plt.subplots(len(venta_productos.columns), figsize=(12, 10), sharex=True) #sharex: comparten el mismo eje x

# Itera a travs de las columnas y traza cada columna en su propia subtrama
for i, column in enumerate(venta_productos.columns):
    if column != 'month_number':
        axs[i].plot(venta_productos.index, venta_productos[column], label=column)
        axs[i].set_ylabel(f'{column} cantidad')
        axs[i].grid(True)
        axs[i].legend()

# Ajusta el espacio entre las subtramas
plt.tight_layout()

plt.show()

```

```python
vacunacion = pd.read_csv('/content/covid-vaccination-doses-per-capita.csv')
vacunacion
```

```python
vacunacion_argentina = vacunacion[vacunacion['Entity'] == 'Argentina'][:]
vacunacion_argentina = vacunacion_argentina.drop('Entity', axis=1)
vacunacion_argentina = vacunacion_argentina.drop('Code', axis=1)
vacunacion_argentina

```

```python

vacunacion_argentina.Day = pd.to_datetime(vacunacion_argentina.Day)

inicio1 = pd.to_datetime('2020-12-29')
fin1 = pd.to_datetime('2021-03-31')
porcion1 = vacunacion_argentina[(vacunacion_argentina.Day >= inicio1) & (vacunacion_argentina.Day <= fin1)]
promedio1 = porcion1['total_vaccinations_per_hundred'].mean()

inicio2 = pd.to_datetime('2021-04-01')
fin2 = pd.to_datetime('2021-06-30')
porcion2 = vacunacion_argentina[(vacunacion_argentina.Day >= inicio2) & (vacunacion_argentina.Day <= fin2)]
promedio2 = porcion2['total_vaccinations_per_hundred'].mean()

df_vacunacion_argentina = pd.DataFrame({'Porcin de fechas': ['1er trimestre 2021', '2do trimestre 2021'],
                                'Promedio de vacunacin': [promedio1, promedio2]})
df_vacunacion_argentina
```

```python

plt.figure(figsize=(8, 4))

plt.plot(df_vacunacion_argentina['Porcin de fechas'], df_vacunacion_argentina['Promedio de vacunacin'])

plt.xlabel('Trimestre')
plt.ylabel('Promedio de vacunacin')
plt.title('Promedio de Vacunacin en Argentina por Trimestre')


plt.grid(True)
plt.show()
```
